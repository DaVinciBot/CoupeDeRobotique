```python
from shapely import *
import matplotlib.pyplot as plt
import geopandas as gpd


def show(e):
    p = gpd.GeoSeries(e)
    p.plot()
    plt.show()
```

```python
l = LineString([(0, 0), (0, 1), (1, 1)])

a = l.buffer(0.5, cap_style=BufferCapStyle.square, join_style=BufferJoinStyle.round)

show(a)
```
